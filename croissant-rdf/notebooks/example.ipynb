```python
import rdflib

g = rdflib.Graph()
g.parse("huggingface.ttl")
test_query = """
SELECT DISTINCT ?a ?b ?c WHERE {?a ?b ?c}"""

qres = g.query(test_query)
for row in qres:
    print(f"{row.b}")
```
