# Use Huggingface Croissant ML via SPARQL

Downloads a turtle file generated by [huggingface-rdf](https://github.com/david4096/huggingface-rdf) and sets up the environment so you can start write queries against it, or add other graphs :)

```python
import rdflib

g = rdflib.Graph()
g.parse("data/huggingface.ttl")
test_query = """
SELECT DISTINCT ?b WHERE {?a ?b ?c}"""

qres = g.query(test_query)
for row in qres:
    print(f"{row.b}")
```
